```python
import numpy as np
import matplotlib.pyplot as plt

gxt_data = np.loadtxt('plots/ground_xt.txt')
nxt_data = np.loadtxt('plots/noisy_xt.txt')
mxt_data = np.loadtxt('plots/mean_xt.txt')
gxt_data = gxt_data[1:]
nxt_data = nxt_data[1:]
start_gt = gxt_data[0, 0]
start_nt = nxt_data[0, 0]
start_mt = mxt_data[0, 0]

gx1 = gxt_data[:, 1]
gx2 = gxt_data[:, 2]
gx3 = gxt_data[:, 3]
gt = gxt_data[:, 0] - start_gt
nx1 = nxt_data[:, 1]
nx2 = nxt_data[:, 2]
nx3 = nxt_data[:, 3]
nt = nxt_data[:, 0] - start_nt
mx1 = mxt_data[:, 1]
mx2 = mxt_data[:, 2]
mx3 = mxt_data[:, 3]
mt = mxt_data[:, 0] - start_mt

plt.plot(gt, gx1, 'seagreen', label='ground')
plt.plot(gt, gx2, 'seagreen')
plt.plot(gt, gx3, 'seagreen')
plt.plot(nt, nx1, 'orange', label='noisy', marker='x', markersize=5)
plt.plot(nt, nx2, 'orange', marker='x', markersize=5)
plt.plot(nt, nx3, 'orange', marker='x', markersize=5)
plt.plot(mt, mx1, 'royalblue', label='weighted mean')
plt.plot(mt, mx2, 'royalblue')
plt.plot(mt, mx3, 'royalblue')
plt.xlabel('time')
plt.ylabel('x')
plt.legend(loc='upper right') 
plt.title('x-time plot')
plt.show()
```

```python
gyt_data = np.loadtxt('plots/ground_yt.txt')
nyt_data = np.loadtxt('plots/noisy_yt.txt')
myt_data = np.loadtxt('plots/mean_yt.txt')
gyt_data = gyt_data[1:]
nyt_data = nyt_data[1:]
start_gt = gyt_data[0, 0]
start_nt = nyt_data[0, 0]
start_mt = myt_data[0, 0]

gy1 = gyt_data[:, 1]
gy2 = gyt_data[:, 2]
gy3 = gyt_data[:, 3]
gt = gyt_data[:, 0] - start_gt
ny1 = nyt_data[:, 1]
ny2 = nyt_data[:, 2]
ny3 = nyt_data[:, 3]
nt = nyt_data[:, 0] - start_nt
my1 = myt_data[:, 1]
my2 = myt_data[:, 2]
my3 = myt_data[:, 3]
mt = myt_data[:, 0] - start_mt

plt.plot(gt, gy1, 'seagreen', label='ground')
plt.plot(gt, gy2, 'seagreen')
plt.plot(gt, gy3, 'seagreen')
plt.plot(nt, ny1, 'orange', label='noisy', marker='x', markersize=5)
plt.plot(nt, ny2, 'orange', marker='x', markersize=5)
plt.plot(nt, ny3, 'orange', marker='x', markersize=5)
plt.plot(mt, my1, 'royalblue', label='weighted mean')
plt.plot(mt, my2, 'royalblue')
plt.plot(mt, my3, 'royalblue')
plt.xlabel('time')
plt.ylabel('y')
plt.legend(loc='upper right') 
plt.title('y-time plot')
plt.show()
```
